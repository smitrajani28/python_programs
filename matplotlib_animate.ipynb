```python
import time
import psutil
import matplotlib.pyplot as plt
```

```python
%matplotlib notebook 
plt.rcParams['animation.html'] = 'jshtml'
```

```python
i = 0
x,y =  [],[]
while True:
    fig = plt.figure()
    ax = fig.subplots()
    x.append(i)
    y.append(psutil.cpu_percent())
    ax.plot(x,y,color='b')
    fig.canvas.draw()
    time.sleep(1)
    i+=1
    fig.show()
```
