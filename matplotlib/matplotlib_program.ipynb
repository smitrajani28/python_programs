```python
from matplotlib import pyplot as plt
from matplotlib.collections import LineCollection
import numpy as np
import random
```

Barchart

```python
x=[1,2,3,4]
y=[2,4,6,8]
plt.xlabel("x",fontsize =20)
plt.ylabel("2x",fontsize=20)
plt.title("y=2x",fontsize=50)
c = ['r','g','b','y']
plt.bar(x,y,width=0.5,color=c,align='edge',edgecolor = 'black',linewidth=5,linestyle=':',alpha=0.5,label='y=2x')
plt.legend()
plt.show()
```

```python
x=['smit','karan','pratik','rahul']
y=[20,46,68,84]
z=[61,44,94,16]

width=0.2
p = np.arange(len(x))
p1= [i+width for i in p]

plt.xlabel("x",fontsize =20)
plt.ylabel("2x",fontsize=20)
plt.title("y=2x",fontsize=50)
c = ['r','g','b','y']
plt.bar(p1,z,width,color='yellow',label='y=2X')
plt.bar(p,y,width,color='red',label='y=2x')
plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```

```python
x=['smit','karan','pratik','rahul']
y=[20,46,68,84]
z=[61,44,94,16]

width=0.2
p = np.arange(len(x))
p1= [i+width for i in p]

plt.xlabel("x",fontsize =20)
plt.ylabel("2x",fontsize=20)
plt.title("y=2x",fontsize=50)
c = ['r','g','b','y']
plt.barh(p1,z,width,color='yellow',label='y=2X')
plt.barh(p,y,width,color='red',label='y=2x')
plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```

scatter

```python
x=[1,2,3,4,5,6,7,8,9]
y=[2,1,4,3,6,5,8,7,8]
colors=['r','y','b','g','y','r','y','b','g']
colors2=[54,87,63,25,15,98,96,45,23]
sizes=[100,600,700,500,9870,500,123,555,444]
plt.scatter(x,y,c=colors2,cmap='viridis',s=sizes,marker="*",edgecolors='g')
plt.colorbar().set_label("colorBar")
plt.title("scatter")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
x=[10,20,30,40]
y=["c","c++","python","java"]
ex=[0,0,0.2,0]
plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%",shadow=True,radius=0.5,labeldistance=1.2,
        startangle=180,textprops={"fontsize":10},counterclock=False,
        wedgeprops={"linewidth":2,"edgecolor" : "m"},
        center=(0,0),rotatelabels=True)
plt.title("pie chart")
plt.legend(loc=3)
plt.show()
```

area or Stack plot

```python
x=[1,2,3,4,5,6,7]
y=[1,3,5,7,9,11,13]
y2=[1,3,6,7,9,1,13]
y3=[1,3,5,8,9,1,13]
l=["c","python","java"]
plt.stackplot(x,y,y2,y3,labels=l,baseline="wiggle")
plt.legend(loc=2)
plt.title("area or stack plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.show()
```

step plot

```python
x=[1,2,3,4,5,6,7]
y=[1,3,5,2,9,1,13]
plt.step(x,y,marker="o",color="r",ms=10,mfc="y",label="python")
plt.legend(loc=2)
plt.title("step plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.show()
```

fill between plot

```python
x=np.array([1,2,3,4,5,6,7])
y=np.array([1,3,5,2,9,1,13])
plt.plot(x,y,label="python")
plt.legend(loc=2)
plt.title("fill between plot")
plt.fill_between(x,y,where=(x>=2) & (x<=4))
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.show()
```

subplot & savefig function

```python
plt.subplot(2,2,1)
x=np.array([1,2,3,4,5,6,7])
y=np.array([1,3,5,2,9,1,13])
plt.plot(x,y,label="python")
plt.legend(loc=2)
plt.title("fill between plot")
plt.fill_between(x,y,where=(x>=2) & (x<=4))
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.subplot(2,2,2)
x=[10,20,30,40]
y=[30,91,54,40]
plt.stem(x,y,linefmt="-.",markerfmt="r+",bottom=45,basefmt="g",label="python",orientation="horizontal")
plt.legend()

plt.subplot(2,2,3)
x=[10,20,30,40]
y=["c","c++","python","java"]
plt.pie(x,labels=y,radius=1.5)
cr=plt.Circle(xy=(0,0),radius=1,facecolor="w")
plt.gca().add_artist(cr)
plt.title("pie chart")

plt.subplot(2,2,4)
x=[10,20,30,40]
y=["c","c++","python","java"]
ex=[0,0,0.2,0]
plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%",shadow=True,radius=0.5,labeldistance=1.2,
        startangle=180,textprops={"fontsize":10},counterclock=False,
        wedgeprops={"linewidth":2,"edgecolor" : "m"},
        center=(0,0),rotatelabels=True)
plt.title("pie chart")

plt.savefig("graphs",dpi=2000,facecolor="y",transparent=True,bbox_inches="tight")

plt.grid()
plt.show()
```

working with text

```python
x=[1,2,3,4]
y=[2,1,4,0]

plt.plot(x,y)
plt.title("hello")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.text(3.2,3.8,"hello, how are you?",bbox={"facecolor":"y"})
plt.annotate("hii there!",xy=(3,2),xytext=(4,0.5),arrowprops=dict(facecolor="black",shrink=100))
plt.show()
```
