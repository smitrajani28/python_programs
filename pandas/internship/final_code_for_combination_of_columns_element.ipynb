```python
import pandas as pd
import openpyxl
from itertools import product
```

```python
def get_color(cell):
    if cell.fill.start_color.index :
        return cell.fill.start_color.index
    return None
```

```python
def get_colored_column_values(color, column, style=None):
        if column == 'CENTER STONE':
            if color == 'FF00FF00':  # Green
                # print("green")
                return ['Moissanite', 'Lab Diamond', 'Natural Diamond', 'Natural Champagne Diamond',
                        'Natural Cornflower Blue Sapphire', 'Natural Emerald', 'Natural Ruby']
            elif color == 'FFFFFF00':  # Yellow
                # print("yellow")
                return ['Moissanite', 'Lab Diamond', 'Natural Diamond']
        elif column == 'BAND STYLE':
            if color == 'FFFF9900':  # Orange
                # print("orange")
                return ['Pave Diamond Sides', 'Plain Band']
        elif column == 'PROFILE':
            if color == 'FF00FFFF':  # Sky Blue
                # print("blue")
                return ['*Only Raised']
            elif color == 'FFFF00FF':  # Purple
                # print("purple")
                return ['*Only Inline']
        return None
```

```python
def get_key(set1,target):
    for key,value in set1.items():
        if target in value:
            return key
```

```python
file_path = 'Copy of Berlinger x Gemist - Product Sheet V1 -PARTNER.xlsx'
wb = openpyxl.load_workbook(file_path)
sheet = wb.active
```

```python
headers = ['SETTING', 'STYLE', 'METAL', 'CENTER STONE', 'BAND STYLE', 'PROFILE']
all_values = {header: set() for header in headers[2:]}
colored_values = {}
heading_columns = {}
current_setting = None
for row in sheet.iter_rows(min_row=2, max_col=len(headers), values_only=False):
    setting = row[1].value
    if setting :
        current_setting = setting
        if current_setting not in colored_values:
            colored_values[current_setting] = {header: set() for header in headers[2:]}
        if row[0].value:
            new_head = row[0].value
            if new_head not in heading_columns:
                heading_columns[new_head] = set()
        
        heading_columns[new_head].add(setting)

    if current_setting:
        for index,cell in enumerate(row[2:],2):
            value = cell.value
            color = get_color(cell)
            values = get_colored_column_values(color,headers[index])
            if values:
                colored_values[current_setting][headers[index]].update(values)
            if value:
                # print(value)
                all_values[headers[index]].update([value])
                # print(headers[index])
pd.DataFrame(colored_values)[:][:]

# print(colored_values)
```

```python
all_combinations = []
for setting, data in colored_values.items():
    # print(setting)
    print(setting)
    setting_combinations = list(product(
        [setting],
        # data['STYLE'] ,
        data['METAL'] ,
        data['CENTER STONE'],
        data['BAND STYLE'] ,
        data['PROFILE']
    ))
    all_combinations.extend(setting_combinations)
```

```python
combinations_df = pd.DataFrame(all_combinations,columns =headers[1:])
combinations_df[headers[0]] = combinations_df[headers[1]].apply(lambda x: get_key(heading_columns,x))
cols = combinations_df.columns.tolist()
cols = cols[-1:] + cols[:-1]
combinations_df2 = combinations_df[cols]

# Remove rows where all columns except SETTING are empty
combinations_df = combinations_df.loc[combinations_df.iloc[:, 1:].any(axis=1)]
```
